```python
import re

data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Сумма: $150.50, 20 EUR, 500 RUB, €75.
"""
pattern_email = r'\w+@\w+\.[a-z]{2,3}'
emails = re.findall(pattern_email, data_text)
print(emails)
```

```python
import re

data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Сумма: $150.50, 20 EUR, 500 RUB, €75.
"""
pattern_date = r'(\d{2})\.(\d{2})\.(\d{4})'
replace = re.sub(pattern_date, r'\3-\2-\1', data_text)
print(replace)
```

```python
import re

data_text = """
Посетите наши страницы:
Email: info@example.com (главный), support-team@corp.net (поддержка)
Дата: 25-05-2023, 01/01/2024, 12.02.2025.
Коды товаров: ID_PROD-1001, ITEM-20-A, ID-300-B.
Сумма: $150.50, 20 EUR, 500 RUB, €75.
"""

pattern = r'(\d+)\s+(EUR|RUB)'

money = re.findall(pattern, data_text)

print()
for amount, currency in money:
    print(f"{amount} {currency}")
```
